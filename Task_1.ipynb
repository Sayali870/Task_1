```python
# Importing libraries
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Loading the dataset
df = pd.read_csv(r"C:\Users\PARTH\Documents\codsoft\task1\titanic-dataset.csv")
df.head()
```

```python
# Plotting the count of survivors with respect to passenger class using matplotlib
plt.figure(figsize=(10, 6))
survived_counts = df[df['Survived'] == 1]['Pclass'].value_counts().sort_index()
not_survived_counts = df[df['Survived'] == 0]['Pclass'].value_counts().sort_index()

bar_width = 0.35
index = survived_counts.index

bar1 = plt.bar(index - bar_width/2, survived_counts, bar_width, label='Survived', color='skyblue')
bar2 = plt.bar(index + bar_width/2, not_survived_counts, bar_width, label='Not Survived', color='salmon')

plt.xlabel('Passenger Class')
plt.ylabel('Count')
plt.title('Survival Count by Passenger Class')
plt.xticks(index)
plt.legend()
plt.show()
```
